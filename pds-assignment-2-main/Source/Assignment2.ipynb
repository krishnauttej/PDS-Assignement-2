```python
import pandas as pd
```

```python
df= pd.read_csv('/content/train.csv')
```

```python
df.head()
```

```python
df.isna().sum() # check for null values
```

```python
d2= pd.DataFrame(df)
```

```python
d2['Mileage'] = d2['Mileage'].str.extract('(\d+\.\d+)').astype(float)

d2['Engine'] = d2['Engine'].str.replace(' CC', '').astype(float)

d2['Power'] = d2['Power'].str.extract('(\d+\.\d+)').astype(float)

d2['New_Price'] = d2['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
d2.head()
```

```python
d2['Mileage'].fillna(d2['Mileage'].median(),inplace=True) # Impute missing numerical values with median
d2['Engine'].fillna(d2['Engine'].median(),inplace=True)
d2['Power'].fillna(d2['Power'].median(),inplace=True)

d2['Seats'].fillna(d2['Seats'].mode()[0],inplace=True) # Impute missing categorical values with mode
```

```python
d2.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column
```

```python
output_csv_file = 'preprocessed_data.csv'
d2.to_csv(output_csv_file, index=False)
```

```python
from datetime import datetime
# Assuming the current year is 2024
current_year = datetime.now().year

# Adding a new feature - Age of the car at the time of purchase
df['Age_at_Purchase'] = current_year - df['Year']
```

```python
#Save the new column addition in separate csv file
output_csv_file = 'column_added.csv'
df.to_csv(output_csv_file, index=False)
```

```python
df.head()
```

```python
#Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset
selected_columns = df[['Name', 'Location', 'Year', 'Kilometers_Driven', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Price']]

# Filtering data based on conditions
filtered_data = df[(df['Year'] >= 2015) & (df['Fuel_Type'] == 'Diesel')]

# Renaming columns
renamed_df = df.rename(columns={'Name': 'Model', 'Kilometers_Driven': 'Odometer', 'Seats': 'Number_of_Seats'})

# Mutating data (adding a new column)
renamed_df['Price_Per_Km'] = renamed_df['Price'] / renamed_df['Mileage'].astype(float)

# Arranging data (sorting)
sorted_data = renamed_df.sort_values(by='Year', ascending=False)

# Summarizing data with group by operations
summary_data = renamed_df.groupby('Model').agg({'Price': 'mean', 'Mileage': 'median'})

# Printing the results
print("\nSelected Columns:")
print(selected_columns.head())

print("\nFiltered Data:")
print(filtered_data.head())

print("\nRenamed DataFrame:")
print(renamed_df[['Model', 'Year', 'Mileage', 'Price', 'Price_Per_Km']].head())

print("\nSorted Data:")
print(sorted_data.head())

print("\nSummary Data:")
print(summary_data.head())
```
